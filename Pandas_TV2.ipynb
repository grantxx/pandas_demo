```python
# Import yfinance
import yfinance as yf  
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
from matplotlib.dates import MO, TU, WE, TH, FR
import numpy as np
import seaborn as seabornInstance 
%matplotlib inline
pd.plotting.register_matplotlib_converters()


Ins_Name = "EURUSD=X"
df = yf.download(Ins_Name,'2019-09-01','2019-12-01')
#df.index = df.index.strftime('%d%m%Y')
df.describe()
fig, ax = plt.subplots(figsize=(15,7))
#ax.plot(X, y, color='red')
plt.scatter(df.index, df.Close, color='crimson', marker='x')
ax.tick_params(axis='x', colors='white', rotation=90)
ax.tick_params(axis='y', colors='white')
#ax.set_xlim(df.index[0] - pd.offsets.Hour(12), df.index[-1]+ pd.offsets.Hour(12))
ax.set_ylabel('Price', fontsize=25)
ax.set_xlabel('Interval', fontsize=25)
ax.set_title('Mean reversion of ' + Ins_Name + ' Close Prices',fontsize= 35,color='white')
ax.xaxis.label.set_color('white')
ax.yaxis.label.set_color('white')
#set ticks every week
ax.xaxis.set_major_locator(mdates.WeekdayLocator(byweekday=(MO, TU, WE, TH, FR)))
# set major ticks format
ax.xaxis.set_major_formatter(mdates.DateFormatter('%b %d'))
# automatically set the padding to fit everything
plt.tight_layout()
plt.show()
```
